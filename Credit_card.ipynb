* Which city has the maximum number of spenders, each month? Is there a need to run campaigns in specific cities?

```python
df12 = df1.merge(df2, on = 'Customer', how='inner')
df12
```

```python
monthly_spenders = df12.groupby(["MonthYear", 'City'])['Customer'].nunique().reset_index()
monthly_spenders
```

```python
monthly_spenders.sort_values("Customer", ascending= False)
```

```python
max_spenders_by_month = monthly_spenders.groupby(['MonthYear'])['Customer'].idxmax()
max_spenders_by_month.sort_values(ascending= False)
```

* Which age group spends the most?

```python
df
```

```python
age_group_spending = df.groupby('Age')['Spend_Amount'].sum().reset_index()
age_group_spending.sort_values("Spend_Amount", ascending= False)
```

```python
max_spending_age_group = age_group_spending.loc[age_group_spending['Spend_Amount'].idxmax()]
max_spending_age_group
```
