# Q1-
* Plot the providers vs number of all wi-fi hotspots they have in operation? Who has the highest installations?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Provider', order=df['Provider'].value_counts(ascending=True).index)
plt.xticks(rotation=90)
plt.title("Number of Wi-Fi Hotspots by Providers")
plt.xlabel("Providers")
plt.ylabel("Number of Hotspots")
plt.show()
```

```python
df.Provider.value_counts()
```

# "LinkNYC - Citybridge" has provided highest number of Wi-Fi connections i.e, 1134.

# Q2-
* Plot providers vs number of free wifi hotspots.

```python
free_provider= df[df["Type"] == "Free"]
all_free_provider= df[(df["Type"] == "Free") | (df["Type"] == "Limited Free")]
Limited_free_provider= df[df["Type"] == "Limited Free"]
free_provider
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data= free_provider, x= "Provider", order=df['Provider'].value_counts(ascending=True).index)
plt.xticks(rotation=90)
plt.title("Number of free Wi-Fi Hotspots by Providers")
plt.xlabel("Providers")
plt.ylabel("Number of free Wi-Fi Hotspots")
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data= all_free_provider, x= "Provider")
plt.xticks(rotation=90)
plt.title("Number of all free Wi-Fi Hotspots by Providers")
plt.xlabel("Providers")
plt.ylabel("Number of free Wi-Fi Hotspots")
plt.show()
```
